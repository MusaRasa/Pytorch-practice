### Check for GPU access with PyTorch


```python
import torch as tr
```

```python
# check if Cuda is avilable
tr.cuda.is_available()
```

For Pytorch since it's capable of running compute on the GPU or CPU, it's best practice to setup device agnostic code: https://pytorch.org/docs/stable/notes/cuda.html#best-practice

```python
device = "cuda" if tr.cuda.is_available() else "cpu"
device
```

### Setting up device agnostic code and putting tensors on and off the GPU

the reason we want our tensor/models on the GPU is because using a GPU result in faster computations.

```python
# Create a tensor (Default on the CPU)
tensor = tr.tensor([2,3,4],device = "cpu")
# Tensor not in GPU
print(tensor, tensor.device)
```

```python
### Move tensor to gpu (if available)
tensor_on_gpu= tensor.to(device)
tensor_on_gpu, 'Not available GPU'
```
