### Pytorch tensor & NumPy
Numpy is a ppular scientific Python numerical computing library.
And because of this, pyTorch has functionlity to interact with it.
* Data in Numpy, want in  PyTorch tensor -> `torch.from_numpy(ndarray)`

```python
# NumPy array to tensor
import numpy as np
import torch as tr
```

```python

array = np.arange(1.0,8.0)
tensor = tr.from_numpy(array).type(tr.float32)
array,tensor
```

```python
array.dtype
```

```python
tensor.dtype
```

```python
 #  change the value of array, what will this do to `tensor`?
array= array+1
array,tensor
```

```python
# Tensor to NumPy array
tensor = tr.ones(7)
numpy_tensor = tensor.numpy()
tensor,numpy_tensor
```

# Changing the tensor, waht happens to `Tensor_numpy`

```python
tensor = tensor+1
tensor,numpy_tensor
```

### `Random_Seed`
In ML and KNN, many of operation as random

```python
# Let's make some random but reproducible tensors
Random_seed = 42
tr.manual_seed(Random_seed)
random_tensor_seed_C = tr.rand(5,2)
tr.manual_seed(Random_seed)
random_tensor_seed_D = tr.rand(5,2)
```

```python
print(random_tensor_seed_C)
```

```python
print(random_tensor_seed_D)
```

```python
print(random_tensor_seed_C == random_tensor_seed_D)
```
